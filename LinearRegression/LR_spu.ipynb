```python
import secretflow as sf

sf.shutdown()
sf.init(parties=['alice','bob'], address='local')
alice, bob = sf.PYU('alice'), sf.PYU('bob')
spu_device = sf.SPU(sf.utils.testing.cluster_def(['alice', 'bob']))
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

X = np.linspace(0,1,100)

y = (np.sin(5 * X * np.pi) * np.exp(-4 * X)) / pow(10,0)

plt.scatter(X,y)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(8192, input_shape = (1,), activation=tf.nn.relu6),
    tf.keras.layers.Dense(1),
])

model.compile(optimizer='adam', loss='mse')
```

```python
def runloop(x,y):
    ep = 5000
    model.fit(x,y,epochs=ep)

dev = spu_device(runloop)(X,y)

# history = model.fit(X,y,epochs=5000)
```
